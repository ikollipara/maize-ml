# Genotype Report

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import tensorflow as tf
import seaborn as sns
```

```python
history = np.load("/home/cse479/ikollipara/genotypes/history.npy", allow_pickle=True)
encoder = tf.keras.models.load_model("/home/cse479/ikollipara/genotypes/encoder.keras")
model = tf.keras.models.load_model("/home/cse479/ikollipara/genotypes/model.keras")
scores = pd.read_csv("/home/cse479/ikollipara/genotypes/scores.csv").rename(index={0: 'loss', 1: 'accuracy'}).drop(columns=['Unnamed: 0'])
```

```python
plt.figure(figsize=(12,8))
plt.suptitle("Training Results for Genotype Auto Encoder")
plt.subplot(1,2,1)
plt.title("Training and Validation Loss")
plt.plot(history.item()['loss'], label="loss")
plt.plot(history.item()['val_loss'], label="val_loss")
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.legend()

plt.subplot(1,2,2)
plt.title("Training and Validation Accuracy")
plt.plot(history.item()['accuracy'], label="accuracy")
plt.plot(history.item()['val_accuracy'], label="val_accuracy")
plt.xlabel("Epoch")
plt.ylabel("Accuracy")
plt.legend()

plt.tight_layout()
plt.show()
```
